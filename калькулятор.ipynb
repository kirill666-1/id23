```python
nums = {
    'один': 1,
    'два': 2,
    'три': 3,
    'четыре': 4,
    'пять': 5,
    'шесть': 6,
    'семь': 7,
    'восемь': 8,
    'девять': 9,
    'десять': 10,
    'одиннадцать': 11,
    'двенадцать': 12,
    'тринадцать': 13,
    'четырнадцать': 14,
    'пятнадцать': 15,
    'шестнадцать': 16,
    'семнадцать': 17,
    'восемьнадцать': 18,
    'девятнадцать': 19,
    'двадцать': 20,
    'тридцать': 30,
    'сорок': 40,
    'пятьдесят': 50,
    'шестьдесят': 60,
    'семьдесят': 70,
    'восемьдесят': 80,
    'девяносто': 90,
    'сто': 100,
    'двести': 200,
    'триста': 300,
    'четыреста': 400,
    'пятьсот': 500,
    'шестьсот': 600,
    'семьсот': 700,
    'восемьсот': 800,
    'девятьсот': 900,
    'одна тысяча': 1000,
    'две тысячи': 2000,
    'три тысячи': 3000,
    'четыре тысячи': 4000,
    'пять тысяч': 5000,
    'шесть тысяч': 6000,
    'семь тысяч': 7000,
    'восемь тысяч': 8000,
    'девять тысяч': 9000
}       #алфавит чисел
symb = {
    'минус':'-',
    'плюс':'+',
    'разделить':'/',
    'умножить':'*'
}   #алфавит символов(не чисел)
seq = input('Введите ваш запрос:').split()
k = 0
while k != len(seq)-1:     #цикл для удаления слова 'на' в списке, для корректной работы программы
    if seq[k] == 'на':
        seq.remove(seq[k])
        continue
    k += 1
def operation(ints):            #функция для подсчета значений
    i = 1
    while i <= len(ints)-1:             
        if ints[i] == 'умножить':
            ints.insert(ints.index('умножить')+2, ints[ints.index('умножить')-1] * ints[ints.index('умножить')+1])
            del ints[ints.index('умножить')-1:ints.index('умножить')+2]
            return operation(ints)
        if ints[i] == 'разделить':
            ints.insert(ints.index('разделить')+2, ints[ints.index('разделить')-1] / ints[ints.index('разделить')+1])
            del ints[ints.index('разделить')-1:ints.index('разделить')+2]
            return operation(ints)
        i += 1
    i = 1
    while i <= len(ints)-1:            
        if ints[i] == 'плюс':
            ints.insert(ints.index('плюс')+2, ints[ints.index('плюс')-1] + ints[ints.index('плюс')+1])
            del ints[ints.index('плюс')-1:ints.index('плюс')+2]
            return operation(ints)
        if ints[i] == 'минус':
            ints.insert(ints.index('минус')+2, ints[ints.index('минус')-1] - ints[ints.index('минус')+1])
            del ints[ints.index('минус')-1:ints.index('минус')+2]
            return operation(ints)
        i += 1
    return ints[0]              

def interp(answer):        
    part, counter = [], 0
    if str(answer)[0] == '-':      
        counter += 1
        part.append('-')
        answer -= (2 * answer)
    if '.' in str(answer):
        inter_answer = int(str(answer[:answer.index('.')]))
        fract_answer = int(str(answer[answer.index('.')+1:]))
    l = len(str(answer))
    while l != 0:
        if l == 2 and str(answer)[-2] == '1':
            part.append(int(str(answer)[len(str(answer)) - l]) * (10 ** (l - 1)) + (int(str(answer)[len(str(answer)) - l+1]) * (10 ** (l - 2))))
            return part    
        part.append(int(str(answer)[len(str(answer)) - l]) * (10 ** (l - 1)))
        l -= 1
    return part
def gk(part):             
    answer_st = ''
    for i in part:
        for j, z in symb.items():
            if i == z:
                answer_st += j + ' '
        for j, z in nums.items():
            if i == z:
                answer_st += j + ' '
    return answer_st
ints, i = [], 0
while i <= len(seq)-2:
    if seq[i] in symb:          
        ints.append(seq[i])
    if seq[i] in nums and seq[i+1] not in nums:  
        ints.append(nums.get(seq[i]))
    if i != len(seq)-1:
        if seq[i] in nums and seq[i+1] in nums:   
            ints.append(nums.get(seq[i])+nums.get(seq[i+1]))
            i += 1
    i +=1
if i != len(seq):          
    if seq[i] in symb:
        ints.append(seq[i])
    if seq[i] in nums:
        ints.append(nums.get(seq[i]))
i = 0
while i <= len(ints)-1:
    if type(ints[i]) is str and ints[i+1] == 'минус':
        ints.insert(i+1,-ints[i+2])
        del ints[i+2:i+4]
    i += 1
print(ints)
print(operation(ints))
print(interp(operation(ints)))
print(gk(interp(operation(ints))))
```
